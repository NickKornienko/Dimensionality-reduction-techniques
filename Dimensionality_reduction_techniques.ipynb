```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE, MDS, Isomap, LocallyLinearEmbedding
from sklearn.datasets import make_classification, make_s_curve
from sklearn.decomposition import TruncatedSVD
from sklearn.manifold import TSNE
import umap
import umap.umap_ as umap
```

```python
# Generate synthetic data
X, y = make_classification(n_samples=1000, n_features=20, n_informative=10, n_redundant=10, random_state=42)

# For non-linear structures
X_s_curve, t = make_s_curve(n_samples=1000, random_state=42)
```

```python
# Applying PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Visualization
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y)
plt.title('PCA Visualization')
plt.show()
```

```python
# Applying SVD
svd = TruncatedSVD(n_components=2)
X_svd = svd.fit_transform(X)

# Visualization
plt.scatter(X_svd[:, 0], X_svd[:, 1], c=y)
plt.title('SVD Visualization')
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.show()
```

```python
# Applying MDS
mds = MDS(n_components=2, random_state=42)
X_mds = mds.fit_transform(X)

# Visualization
plt.scatter(X_mds[:, 0], X_mds[:, 1], c=y)
plt.title('MDS Visualization')
plt.show()
```

```python
# Applying ISOMap
isomap = Isomap(n_components=2)
X_isomap = isomap.fit_transform(X)

# Visualization
plt.scatter(X_isomap[:, 0], X_isomap[:, 1], c=y)
plt.title('ISOMap Visualization')
plt.show()
```

```python
# Applying LLE
lle = LocallyLinearEmbedding(n_components=2)
X_lle = lle.fit_transform(X)

# Visualization
plt.scatter(X_lle[:, 0], X_lle[:, 1], c=y)
plt.title('LLE Visualization')
plt.show()
```

```python
# Applying UMAP
reducer = umap.UMAP(random_state=42)
X_umap = reducer.fit_transform(X)

# Visualization
plt.scatter(X_umap[:, 0], X_umap[:, 1], c=y, cmap='viridis')
plt.title('UMAP Visualization')
plt.show()
```

```python
# Applying t-SNE
tsne = TSNE(n_components=2, perplexity=30, learning_rate=200, random_state=42)
X_tsne = tsne.fit_transform(X)

# Visualization
plt.figure(figsize=(10, 6))
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap='viridis')
plt.colorbar()
plt.title('t-SNE Visualization')
plt.xlabel('t-SNE feature 1')
plt.ylabel('t-SNE feature 2')
plt.show()
```
